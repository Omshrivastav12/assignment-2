```python
#Q1 def ekyword is used to define fun
def odd(l):
    l1=[]
    for i in l:
        if i%2 !=0:
            l1.append(i)
    return l1
odd(list(range(1,26)))
```

```python
#Q4 generator function is the function that generates output continuously till a point we wanted it to be without storing everything in to memory
#yield is used to define outcome of generator funtion eg.
def feb(n):
    a,b = 0,1
    for i in range(n):
        yield a
        a,b=b,a+b
for i in feb(6):
    print(i)
```

```python
#Q6 py6thon for fibonacci series using while loop\
def fib():
    a,b = 0,1
    while True:
        yield a
        a,b=b,a+b
yo=fib()
for i in range(11):
    print(next(yo))
    
```
